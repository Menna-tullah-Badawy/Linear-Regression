Salary Prediction with Machine Learning
(Linear Regression
)


```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv('/content/Experience-Salary.csv')
```

```python
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
Figure = px.scatter(data, x="exp(in months)", y="salary(in thousands)", title="Experience vs Salary", size="exp(in months)", trendline="ols")
Figure.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = np.asanyarray(data[["exp(in months)"]])
y = np.asanyarray(data[["salary(in thousands)"]])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
a = float(input("exp(in months) : "))
features = np.array([[a]])
print("Predicted Salary = ", model.predict(features))
```
